* Data pre-processing and Data Cleaning

```python
# drop the can't used columns and columns haveing above 40% of nullvalues

df.drop(['uniq_id','description','product_information','product_description','customer_questions_and_answers','sellers','customers_who_bought_this_item_also_bought','items_customers_buy_after_viewing_this_item'],axis=1,inplace=True)
df.head(3)
```

### check the every columns and pre-process the data

#### Clean & Transforming the data for Analysis point of view

```python
## check the columns nullvalues and handle the columns
# price column

df['price']=df['price'].str.replace('£','',regex=True)
df["price"] = pd.to_numeric(df["price"], errors="coerce")
df["price"].fillna(df["price"].median(), inplace=True)
df.rename(columns={"price": "price in £"}, inplace=True)

# number_available_in_stock
df["number_available_in_stock"] = pd.to_numeric(df["number_available_in_stock"].astype(str).str.extract("(\d+)")[0], errors="coerce")
df["number_available_in_stock"].fillna(0, inplace=True)

# number_of_reviews
df["number_of_reviews"] = pd.to_numeric(df["number_of_reviews"], errors="coerce").fillna(0).astype(int)

##  number_of_answered_questions
df["number_of_answered_questions"]=df["number_of_answered_questions"].astype(float) #change the datatype
df["number_of_answered_questions"].fillna(0,inplace=True)

# average_review_rating
df["average_review_rating"] = df["average_review_rating"].astype(str).str.extract("(\d+\.?\d*)").astype(float)
df["average_review_rating"].fillna(df["average_review_rating"].median(), inplace=True)
```

```python
## amazon_category_and_sub_category  this column divided into three columns
df[["amazon_category", "sub_category_1", "sub_category_2"]] = df["amazon_category_and_sub_category"].str.split(" > ", n=2, expand=True)

# also drop the this column and the sub_category_2 has above 40% of null values
df.drop(['amazon_category_and_sub_category','sub_category_2'],axis=1,inplace=True)

### filling the null values of manufacturer,amazon_category,sub_category_1

# Fill missing values with "Unknown"
df.loc[:, ["manufacturer", "amazon_category", "sub_category_1"]] = df[
    ["manufacturer", "amazon_category", "sub_category_1"]
].fillna("Unknown")

# Fill missing values with the mode
df.loc[:, ["manufacturer", "amazon_category", "sub_category_1"]] = df[
    ["manufacturer", "amazon_category", "sub_category_1"]
].fillna(df.mode().iloc[0])

# Forward fill remaining missing values
df.loc[:, ["manufacturer", "amazon_category", "sub_category_1"]] = df[
    ["manufacturer", "amazon_category", "sub_category_1"]
].fillna(method="ffill")

# customer_reviews in this column extracting costomer reviews 
df["customer_reviews"] = df["customer_reviews"].str.extract("(\d+)").astype(float)
df["customer_reviews"].fillna(0,inplace=True)
```

#### Remove the outlaiers

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)  # 25th percentile
    Q3 = df[column].quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1  # Interquartile Range

    # Define lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Remove rows where the column value is outside the bounds
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    
    return df
```

```python
# Remove outliers from price and review rating
df = remove_outliers(df, "price in £")
df = remove_outliers(df, "average_review_rating")
```

### MULTI VARIATE ANALYSIS :

### Analyzes the most common main categories and subcategories :

```python
###  Analyzes the most common main categories and subcategories.

def analyze_common_categories(df, top_n=10):
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # Top Main Categories
    top_main = df["amazon_category"].value_counts().head(top_n)
    sns.barplot(x=top_main.values, y=top_main.index, ax=axes[0], palette="viridis")
    axes[0].set_title(f"Top {top_n} Main Categories")
    axes[0].set_xlabel("Number of Products")

    # Top Subcategories
    top_sub = df["sub_category_1"].value_counts().head(top_n)
    sns.barplot(x=top_sub.values, y=top_sub.index, ax=axes[1], palette="magma")
    axes[1].set_title(f"Top {top_n} Subcategories")
    axes[1].set_xlabel("Number of Products")

    plt.tight_layout()
    plt.show()

# Run the analysis
analyze_common_categories(df)
```

```python
### Analyzes the average price of products across top categories

def category_price_analysis(df, top_n=10):
    
    # Compute average price per category
    category_avg_price = df.groupby("amazon_category")["price in £"].mean().sort_values(ascending=False).head(top_n)

    # Plot the results
    plt.figure(figsize=(12, 5))
    sns.barplot(x=category_avg_price.values, y=category_avg_price.index, palette="coolwarm")
    plt.title(f"Top {top_n} Categories by Average Price")
    plt.xlabel("Average Price ($)")
    plt.ylabel("Category")
    plt.show()

# Run the analysis
category_price_analysis(df)
```

```python
## Plots the price distribution for selected main categories.
def price_distribution_by_category(df, selected_categories=5):
    
    # Select top categories based on product count
    top_categories = df["amazon_category"].value_counts().index[:selected_categories]
    df_filtered = df[df["amazon_category"].isin(top_categories)]

    # Create a boxplot to show price distribution
    plt.figure(figsize=(10, 4))
    sns.boxplot(x="amazon_category", y="price in £", data=df_filtered, palette="Set2")
    plt.xticks(rotation=45)
    plt.title("Price Distribution Across Categories")
    plt.xlabel("Category")
    plt.ylabel("Price (£)")
    plt.show()

# Run the analysis
price_distribution_by_category(df)
```
